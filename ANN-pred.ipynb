```python
import tensorflow as tf
import datetime
from tensorflow.keras.models import load_model
import pickle
import numpy as np
import pandas as pd
```

```python
model=load_model('model.h5')
with open('label_encode_gender.pkl','rb') as file:
    label_gender = pickle.load(file)
with open('label_onehot_encoder_geo.pkl','rb') as file:
    label_geo=pickle.load(file)
with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)    
      
```

```python
input_data={
'CreditScore':600,
'Geography':'France',
'Gender':"Male",
'Age':40,
'Tenure':3,
'Balance':6000,
'NumOfProducts':2,
'HasCrCard':1,
'IsActiveMember':1,
'EstimatedSalary':50000
            
}
```

```python
input_data=pd.DataFrame([input_data])
```

```python
geoEncoded=label_geo.transform(input_data[['Geography']]).toarray()
```

```python
df_geoEncoded=pd.DataFrame(geoEncoded,columns=label_geo.get_feature_names_out(['Geography']))
```

```python
genderEncoded=label_gender.transform(input_data[['Gender']])
```

```python
input_data=pd.concat([input_data.drop(['Geography'],axis=1), df_geoEncoded],axis=1)
```

```python
input_data
```

```python
input_data['Gender']=genderEncoded
```

```python
input_data
```

```python
input_data_scalled=scaler.transform(input_data)
input_data_scalled
```

```python
#predict the person leaves the bank or %notebook
get_pred=model.predict(input_data_scalled)
get_pred
```

```python
get_pr=get_pred[0][0]

get_pr
```

```python
if get_pr>0.5:
    print("The person will leave the bank")
else:
    print("The person will not leave the bank")
```
